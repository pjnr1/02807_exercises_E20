```python
import pandas as pd
import numpy as np
```

Load the data set

```python
url = 'https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/adult.data.csv'
df = pd.read_csv(url)
df.head()
```

-----

 1. How many men and women (sex feature) are represented in this dataset?

```python
n_males = df[df["sex"] == "Male"].shape[0]
n_females = df[df["sex"] == "Female"].shape[0]
print(f'Count of males:   {n_males}')
print(f'Count of females: {n_females}')
```

-----
 2. What is the average age (age feature) of women?

```python
avg_female_age = np.mean(df[df["sex"] == "Female"]["age"])
print(f'Average age of females in dataset: {avg_female_age}')
```

-----
 3. What is the percentage of German citizens (native-country feature)?

```python
n_germans = df[df["native-country"] == "Germany"].shape[0]
p_germans = n_germans / df.shape[0] * 100
print(f'Percentage German citizens: {p_germans}')
```

-----
 4. (5.) What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

```python
over_stats = df[df["salary"] == ">50K"]["age"].describe()[["mean", "std"]]
under_stats = df[df["salary"] == "<=50K"]["age"].describe()[["mean", "std"]]
print(f'Age of salary,  > 50K: {round(over_stats[0],2)} +/- {round(over_stats[1],2)}')
print(f'Age of salary, <= 50K: {round(under_stats[0],2)} +/- {round(under_stats[1],2)}')
```

-----
 6. Is it true that people who earn more than 50K have at least high school education? (education = HS-grad, Some-College, Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or *Doctorate)


```python
u_edu = df[df["salary"] == ">50K"]["education"].unique()
higher_educations = [
    "HS-grad",
    "Some-college",
    "Bachelors",
    "Prof-school",
    "Assoc-acdm",
    "Assov-voc",
    "Masters",
    "Doctorate"
]
answer = "Yes"
for high_sal_edu in u_edu:
    if high_sal_edu not in higher_educations:
        answer = "No"
        break
print(f'The answer is "{answer}"')
```

 7. Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.

 8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

```python
male_above_50k = df[(df["salary"] == ">50K") & (df["sex"] == "Male")]
n_married = male_above_50k[(["Married" in x for x in male_above_50k["marital-status"]])].shape[0]
n_bachelors = male_above_50k[(["Married" not in x for x in male_above_50k["marital-status"]])].shape[0]
print(f'Married:   {n_married}')
print(f'Bachelors: {n_bachelors}')
```

 9. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python
print(f'Max hours pr week: {np.max(df["hours-per-week"])}')
print(f'Number of persons: {df[df["hours-per-week"] == 99].shape[0]}')
print(f'Percentage >50K:   {round(df[(df["hours-per-week"] == 99) & (df["salary"] == ">50K")].shape[0] / df[df["hours-per-week"] == 99].shape[0] * 100, 2)}')
```
